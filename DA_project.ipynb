```python
# --- 3. Statistical Analysis ---
print("\n--- Descriptive Statistics ---")
print(df.describe())
```

```python
# Specific analysis on 'unitprice' column
print("\n--- Unit Price Statistics ---")
print("Count:", df['unitprice'].count())
print("Mean:", df['unitprice'].mean())
print("Min:", df['unitprice'].min())
print("Max:", df['unitprice'].max())
print("Median:", df['unitprice'].median())
print("Std:", df['unitprice'].std())
print("Var:", df['unitprice'].var())
```

```python
# --- 4. Data Visualization & Analysis Questions ---

# 1. Best Year of Sales (Number of Sales)
yearly_sales = df.groupby('year')['amount'].sum()

best_year = yearly_sales.idxmax()
print("Best year of sales:", best_year)

yearly_sales.plot(kind='bar')
plt.title("Total Sales by Year")
plt.ylabel("Total Sales")
plt.xlabel("Year")
plt.show()
```

```python
# 2. Best Month of Sales
month_group = df.groupby('month')['item_id'].count()
plt.figure(figsize=(10, 6))
plt.plot(month_group.index, month_group.values, color='green', marker='o')
plt.xlabel('Months')
plt.ylabel('Number of sales')
plt.title('Number of sales happened in each month')
plt.xticks(range(1, 13))
plt.grid(True)
plt.savefig('2_sales_per_month.png')
plt.show()
```

```python
# 3. Total Sales Amount by Year
year_group_sum = df.groupby('year')['amount'].sum()
plt.figure(figsize=(10, 6))
plt.bar(year_group_sum.index, year_group_sum.values, color='green', alpha=0.6)
plt.plot(year_group_sum.index, year_group_sum.values, color='red', marker='o')
plt.xlabel('Years')
plt.ylabel('Total sales')
plt.title('Total sales happened in each year')
plt.savefig('3_total_sales_amount_year.png')
plt.show()
```

```python
# 4. Top 10 Brands (2015-2018)
filter_years = (df['year'] >= 2015) & (df['year'] <= 2018)
df_15_18 = df[filter_years]
brand_group = df_15_18.groupby('brand')['amount'].sum().nlargest(10)

plt.figure(figsize=(12, 6))
plt.bar(brand_group.index, brand_group.values, color='green')
plt.xlabel('Brand')
plt.ylabel('Total sales')
plt.title('Top 10 Brands based on total sales (2015-2018)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('4_top_10_brands.png')
plt.show()
```

```python
# 5. Top 5 Brands for Specific Years (2016, 2017, 2018)
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
years_to_plot = [2016, 2017, 2018]
colors = ['red', 'green', 'blue']

for i, year in enumerate(years_to_plot):
    df_year = df[df['year'] == year]
    top_brands = df_year.groupby('brand')['amount'].sum().nlargest(5)
    axs[i].bar(top_brands.index, top_brands.values, color=colors[i])
    axs[i].set_title(f'Top 5 brands in {year}')
    axs[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('5_yearly_top_brands.png')
plt.show()
```

```python
# 6. Top Categories (2015-2018)
category_sales = df.groupby('category')['quantity'].sum().sort_values(ascending=False)

category_sales.plot(kind='bar')
plt.title("Sales by Product Category")
plt.ylabel("Quantity Sold")
plt.xlabel("Category")
plt.show()
```

```python
#7.Brand that sold Least(2015-2018)
least_brand = (
    df[(df['year'] >= 2015) & (df['year'] <= 2018)]
    .groupby('brand')['quantity']
    .sum()
    .idxmin()
)

print("Least sold brand (20152018):", least_brand)
```

```python
# 8. Distribution of Ratings
rating_counts = df['rating'].value_counts().sort_index()
plt.figure(figsize=(8, 5))
plt.bar(rating_counts.index, rating_counts.values, color='red')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Frequency of different ratings')
plt.savefig('7_rating_distribution.png')
plt.show()
```

```python
# 9. Most Rated Brands (2015-2018) - Average Rating
most_rated_brand = (
    df[(df['year'] >= 2015) & (df['year'] <= 2018)]
    .groupby('brand')['rating']
    .mean()
    .idxmax()
)

print("Most rated brand (20152018):", most_rated_brand)
```

```python
# 10. Category Sales Percentage
category_percentage = (
    df.groupby('category')['amount'].sum() /
    df['amount'].sum()
) * 100

category_percentage.plot(kind='pie', autopct='%1.1f%%')
plt.title("Category-wise Percentage Sales")
plt.ylabel("")
plt.show()
```

```python
# 11. Brand Sales Percentage
brand_pie = df.groupby('brand')['amount'].sum().nlargest(5)
plt.figure(figsize=(8, 8))
plt.pie(brand_pie.values, labels=brand_pie.index, autopct="%.2f%%")
plt.title('Top 5 Brands with highest percentage of sales')
plt.savefig('10_brand_pie.png')
plt.show()
```

```python
# 12. Gender Distribution
gender_dist = df['gender'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(gender_dist.values, labels=gender_dist.index, autopct='%1.1f%%', colors=['pink', 'lightblue'])
plt.title('Gender wise customer Distribution')
plt.savefig('11_gender_distribution.png')
plt.show()
```

```python
# 13. Gender vs Category Preference
gender_category = df.groupby(['gender', 'category'])['quantity'].sum()
genders = df['gender'].unique()

for gender in genders:
    plt.figure(figsize=(8, 8))
    data = gender_category[gender]
    plt.pie(data, labels=data.index, autopct='%.1f%%')
    plt.title(f'{gender} - Product Preferences')
    plt.savefig(f'12_gender_pref_{gender}.png')
    plt.show()
```
